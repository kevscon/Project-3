# Election Data
Collected from Tony McGovern GitHub page.

```python
import pandas as pd
import numpy as np
```

```python
# import 2016 election results by county

df_election = pd.read_csv('2016_US_County_Level_Presidential_Results.csv')
```

```python
df_election.rename(columns={'combined_fips': 'code'}, inplace=True)
```

```python
df_election.head()
```

```python
# import county data

df_county_data = pd.read_csv('county_out_2.csv')
```

```python
df_county_data.head()
```

```python
df = pd.merge(df_county_data, df_election, on='code', how='left')
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# define function to reformat rate values

def clean_rate(rate_str):
    try:
        return float(str(rate_str).replace('%', ''))
    except:
        pass
```

```python
# define function to reformat monetary and count values

def clean_num(num_str):
    try:
        return int(str(num_str).replace(',',''))
    except:
        pass
```

```python
df['diff'] = df['diff'].apply(clean_num)
df['per_point_diff'] = df['per_point_diff'].apply(clean_rate)
```

```python
# output file for EDA

df.to_csv('county_election_data_2.csv')
```
